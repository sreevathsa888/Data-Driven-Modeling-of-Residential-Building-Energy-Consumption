```python
df = pd.read_csv("data/energy_efficiency.csv", encoding='latin1')
df.head()
```

```python
df.columns = [
    "Relative_Compactness",
    "Surface_Area",
    "Wall_Area",
    "Roof_Area",
    "Height",
    "Orientation",
    "Glazing_Area",
    "Glazing_Distribution",
    "Heating_Load",
    "Cooling_Load"
]
```

SEPARATE FEATURES & TARGETS

```python
X = df.drop(["Heating_Load", "Cooling_Load"], axis=1)

y_heating = df["Heating_Load"]
y_cooling = df["Cooling_Load"]
```

TRAIN-TEST SPLIT

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_h_train, y_h_test, y_c_train, y_c_test = train_test_split(
    X, y_heating, y_cooling, test_size=0.2, random_state=42
)
```

FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

VERIFY SHAPES

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

print("Heating train shape:", y_h_train.shape)
print("Cooling train shape:", y_c_train.shape)
```

The dataset was divided into training (80%) and testing (20%) subsets to evaluate model performance. Feature scaling was performed using StandardScaler to normalize input features. This ensures that machine learning algorithms sensitive to feature magnitude, such as Support Vector Regression and K-Nearest Neighbors, perform effectively.
